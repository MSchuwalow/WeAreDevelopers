```python
import matplotlib
matplotlib.use('Agg')
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import imread, imresize
from os.path import join
from keras.models import Sequential, Model
from keras.layers import Dense, Conv2D, Activation, Flatten, Input
from keras import backend as K
from vis.visualization import visualize_cam, visualize_saliency, visualize_activation
from vis.utils import utils
from vis.losses import ActivationMaximization
from vis.regularizers import TotalVariation, LPNorm
from vis.optimizer import Optimizer
from vis.callbacks import GifGenerator
from vis.utils.vggnet import VGG16
import json
import vis.utils
from keras.preprocessing.image import img_to_array
from keras.applications.imagenet_utils import preprocess_input

DATA_HOME = "../images"
OUT_DIR = "../out"
RES_DIR = "../resources"
IMAGE_DIM = 224
```

```python
# Load pretrained VGG-16
model = VGG16(weights='imagenet', include_top=True)
print(model.summary())
```

```python
with open(join(RES_DIR, "imagenet_class_index.json")) as f:
    CLASS_INDICES = json.load(f)
        
def get_imagenet_label(indices, join=', '):
    return [CLASS_INDICES[str(idx)][1] for idx in indices]
```

```python
# Say hello to our good frien
octocat = imread(join(DATA_HOME, "octocat.jpg"))
octocat = imresize(octocat, size=(IMAGE_DIM, IMAGE_DIM))
plt.imshow(octocat)
plt.axis("off")
plt.savefig(join(OUT_DIR, "octocat.svg"))

# Get the first layer from VGG-16 and get the activations
model1 = Sequential()
model1.add(model.layers[0])
model1.add(model.layers[1])
model1.add(model.layers[2])
model1.add(Activation('relu'))

activations = model1.predict(np.asarray([octocat]))
fig, axis = plt.subplots(figsize=(16, 16), nrows=3, ncols=3)
for i, ax in enumerate(axis.flat):
    ax.set_title("Filter %s" % (i + 1))
    ax.imshow(activations[0, :, :, i], cmap="plasma")
    ax.axis("off")
fig.tight_layout()
plt.show()
fig.savefig(join(OUT_DIR,"convolutions.svg"), format="svg")
```

```python
# Get images that maximise the activation in a certain layer
layer_name = 'predictions'
layer_idx = [idx for idx, layer in enumerate(model.layers) if layer.name == layer_name][0]

# Generate three different images of the same output index.
fig, axis = plt.subplots(figsize=(15, 6), nrows=2, ncols=5)
for i, ax in enumerate(axis.flat):
    image = visualize_activation(model, layer_idx, filter_indices=[2*i], text=get_imagenet_label([2*i])[0], max_iter=500)
    ax.imshow(image)
    ax.axis("off")
fig.tight_layout()
plt.show()
fig.savefig(join(OUT_DIR,"activations.svg"), format="svg")
```

```python
# Create activation gif
layer_name = 'predictions'
layer_dict = dict([(layer.name, layer) for layer in model.layers[1:]])
output_class = [283] # cats because reasons

losses = [
    (ActivationMaximization(layer_dict[layer_name], output_class), 2),
    (LPNorm(model.input), 10),
    (TotalVariation(model.input), 10)
]
opt = Optimizer(model.input, losses)
opt.minimize(max_iter=500, callbacks=[GifGenerator(join(OUT_DIR, 'cat_progress'))])
```

```python
# We want a model with softmax here
bottom_model = VGG16(weights='imagenet', include_top=True, input_shape=(IMAGE_DIM, IMAGE_DIM, 3))
weights = bottom_model.layers[-1].get_weights()
bottom_model.layers.pop()
top = Dense(1000, activation="linear")(bottom_model.layers[-1].output)

model2 = Model(input=bottom_model.input, output=[top])
model2.layers[-1].set_weights(weights)

# Images corresponding to tiger, penguin, dumbbell, speedboat, spider
image_paths = [
    join(DATA_HOME, "flamingo.jpg"),
    join(DATA_HOME, "tiger_cat.jpg"),
    join(DATA_HOME, "dumbbell.jpg"),
    join(DATA_HOME, "spider.jpg"),
    join(DATA_HOME, "birdhouse.jpg")
]

fig, axis = plt.subplots(figsize=(16, 16), nrows=1, ncols=5)
for i, ax in enumerate(axis.flat):
    seed_img = utils.load_img(image_paths[i], target_size=(224, 224))
    x = np.expand_dims(img_to_array(seed_img), axis=0)
    x = preprocess_input(x)
    pred_class = np.argmax(model2.predict(x))
    # Here we are asking it to show attention such that prob of `pred_class` is maximized.
    heatmap = visualize_cam(model2, len(model2.layers) -1, [pred_class], seed_img)
    ax.imshow(heatmap)
    ax.set_title(get_imagenet_label([pred_class])[0])
    ax.axis("off")

fig.tight_layout()
plt.show()
fig.savefig(join(OUT_DIR,"heatmap.svg"), format="svg")
```
